#### Understanding File System and Python Organization(How libraries work in Python): Folder, File, Module, and Package

First clarify the distinction between these related terms in the context of Python:

*   **File**: At its core, a file is a unit of data stored on a file system. In Python, a file with a `.py` extension contains Python source code.

*   **Module**: A module in Python is a single `.py` file containing Python code (functions, classes, variables, etc.). When we `import` a module, we make the code defined within that file accessible in our current script. Modules are the basic building blocks for organizing reusable code.

*   **Folder / Directory**: These are terms from the file system, referring to containers that hold files and other folders. In Python, a folder's role is significant when it comes to packages.

*   **Package**: A package is a way to organize related modules into a hierarchical structure using folders. A folder becomes a Python package if it contains a special file named `__init__.py` (which can be empty). This `__init__.py` file signals to Python that the directory should be treated as a package. Packages help in managing larger projects by providing a clear way to structure and namespace modules, preventing naming conflicts.

In essence:

*   A **module** is a single Python **file**.
*   A **package** is a **folder** containing **modules** (and potentially other packages) and an `__init__.py` file.

Think of it like this:
- A single recipe card is a **module** (a file with instructions).
- A recipe box organized by cuisine (with sections for Italian, Mexican, etc., and maybe an index card for each section) is a **package** (a folder with modules/recipes and an `__init__.py` equivalent).
- The physical pieces of paper are the **files**.
- The container (box) is the **folder/directory**.

##### Understanding NLTK Structure: Modules, Classes, and Objects

First clarify the concepts of modules, classes, and objects as they apply to working with NLTK corpora like `gutenberg`.

In Python:

*   **Module**: A module is a file containing Python code. It can define functions, classes, and variables. Modules help organize code into logical units. When you `import module_name`, you make the contents of that module available.
*   **Class**: A class is a blueprint for creating objects. It defines the attributes (data) and methods (functions) that objects created from that class will have. Think of it as a template.
*   **Object**: An object is an instance of a class. It's a concrete entity created from the class blueprint, with its own unique data (based on the class's attributes) and the ability to perform actions (using the class's methods).

Now, let's relate these to NLTK:

This `gutenberg` object is an **instance** of a `GutenbergCorpusReader` class (or a similar subclass). #### Basic Usage: Tokenization and Stemming

**Tokenization** is the process of breaking down a text into individual words or tokens.
**Stemming** is the process of reducing words to their root or base form.

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer

sample_text = "NLTK is a powerful library for natural language processing. It provides tools for tokenization, stemming, tagging, parsing, and more."

# Tokenization
tokens = word_tokenize(sample_text)
print("Original Text:", sample_text)
print("Tokens:", tokens)

# Stemming
stemmer = PorterStemmer()
stemmed_tokens = [stemmer.stem(token) for token in tokens]
print("\nStemmed Tokens:", stemmed_tokens)
```

#### Tagging and Parsing

Let's explore Part-of-Speech (POS) tagging and simple parsing.

**Part-of-Speech Tagging** is the process of assigning a grammatical category (like noun, verb, adjective, etc.) to each word in a sentence.

**Parsing** involves analyzing the grammatical structure of a sentence to determine the relationships between words.

#### Part-of-Speech Tagging

This cell performs Part-of-Speech tagging on the `sample_text`. It uses the `word_tokenize` function to break the text into tokens and then `nltk.pos_tag` to assign a tag to each token.

```python
import nltk
from nltk.tokenize import word_tokenize

sample_text = "NLTK is a powerful library for natural language processing. It provides tools for tokenization, stemming, tagging, parsing, and more."

# Part-of-Speech Tagging
tokens = word_tokenize(sample_text)
pos_tags = nltk.pos_tag(tokens)

print("Part-of-Speech Tagging:")
print(pos_tags)
```

#### Simple Parsing

This cell demonstrates simple parsing using a regular expression parser and a predefined grammar. The grammar defines a simple Noun Phrase (NP) as an optional determiner (`DT`), followed by zero or more adjectives (`JJ`), and ending with a noun (`NN`). The parser then applies this grammar to the POS-tagged tokens to find noun phrases.

```python
import nltk
from nltk.tokenize import word_tokenize

# Simple Parsing (using a predefined grammar)

grammar = "NP: {<DT>?<JJ>*<NN>}"  # Noun Phrase grammar
cp = nltk.RegexpParser(grammar)
result = cp.parse(pos_tags)

print("\nSimple Parsing:")
print(result)

# To visualize the parse tree (requires installing graphviz and ghostscript)
# result.draw()
```

#### Sentiment Analysis

This section introduces basic sentiment analysis using NLTK's VADER (Valence Aware Dictionary and sEntiment Reasoner) lexicon. A **lexicon** in this context is essentially a dictionary or vocabulary, often with associated information for each word. VADER's lexicon contains words and their associated sentiment intensity scores.

VADER is a lexicon and rule-based sentiment analysis tool that is specifically attuned to sentiments expressed in social media, but it also works remarkably well on general text. It doesn't require any training data and can provide sentiment scores directly.

#### Using VADER for Sentiment Analysis

This cell demonstrates how to use VADER to get sentiment scores for a given sentence.

1.  **Import necessary modules**: We import `SentimentIntensityAnalyzer` from `nltk.sentiment.vader`.
2.  **Initialize the Analyzer**: An instance of `SentimentIntensityAnalyzer` is created. This object contains the VADER lexicon and rules.
3.  **Define a sample sentence**: A string variable `sentence` is created to hold the text we want to analyze.
4.  **Get Polarity Scores**: The `polarity_scores()` method of the analyzer is called with the sentence as input. This method returns a dictionary containing four scores:
    *   `neg`: The probability of the sentence being negative.
    *   `neu`: The probability of the sentence being neutral.
    *   `pos`: The probability of the sentence being positive.
    *   `compound`: A normalized, weighted composite score that summarizes the sentiment. This score is often the most useful single metric as it accounts for the intensity and polarity of the sentiment. It ranges from -1 (most negative) to +1 (most positive).
5.  **Print Scores**: The sentence and the calculated sentiment scores are printed.
6.  **Interpret Compound Score**: The code then provides a simple interpretation of the `compound` score:
    *   If `compound_score` is 0.05 or greater, the sentiment is considered Positive.
    *   If `compound_score` is -0.05 or less, the sentiment is considered Negative.
    *   Otherwise, the sentiment is considered Neutral.

This interpretation provides a quick way to categorize the overall sentiment of the sentence based on the compound score.

```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer


# Initialize the VADER sentiment intensity analyzer
analyzer = SentimentIntensityAnalyzer()

# Sample sentence for sentiment analysis
sentence = "NLTK is a powerful library and I really enjoy using it!"

# Get the sentiment scores
sentiment_scores = analyzer.polarity_scores(sentence)

print(f"Sentence: {sentence}")
print(f"Sentiment Scores: {sentiment_scores}")

# Interpret the compound score
compound_score = sentiment_scores['compound']

if compound_score >= 0.05:
    sentiment = 'Positive'
elif compound_score <= -0.05:
    sentiment = 'Negative'
else:
    sentiment = 'Neutral'

print(f"Overall Sentiment: {sentiment}")
```

### Question 3: Top 30 Most Frequent Words in a Negative Movie Review

```python
import nltk
from nltk.corpus import movie_reviews, stopwords, wordnet
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import string
from nltk import FreqDist
import matplotlib.pyplot as plt
import pandas as pd


#Load and Explore Movie Reviews Corpus and Get file IDs for the 'neg' category
neg_file_ids = movie_reviews.fileids('neg')
print("File IDs for Negative Reviews (first 10):")
print(neg_file_ids[:10])


#Select and Load a Negative Review
selected_neg_file_id = neg_file_ids[5]
neg_review_text = movie_reviews.raw(selected_neg_file_id)

print(f"\nSelected Negative Review File ID: {selected_neg_file_id}")
print("\nFirst 500 characters of the review text:")
print(neg_review_text[:500])


# Tokenize the raw text
neg_review_tokens = word_tokenize(neg_review_text)
print("\nFirst 20 tokens of the negative review:")
print(neg_review_tokens[:20])


#Preprocess the Text
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

# Preprocess the tokens: remove punctuation, convert to lowercase, remove stopwords, lemmatize
cleaned_words = []
for token in neg_review_tokens:
    # Remove punctuation attached with words and convert to lowercase
    cleaned_token = token.translate(str.maketrans('', '', string.punctuation)).lower()

    if cleaned_token and cleaned_token not in stop_words:
        lemmatized_word = lemmatizer.lemmatize(cleaned_token) #additional info: lemmatizer can also take pos tags as argument instead of general token.
        cleaned_words.append(lemmatized_word)

print("\nFirst 20 cleaned and lemmatized words:")
print(cleaned_words[:20])


#frequency distribution of the cleaned words
fdist = FreqDist(cleaned_words)
print(fdist)

# top 30 most common words and their frequencies
top_30_words = fdist.most_common(30)
print("\n30 Most Frequent Words and their Frequencies:")
print(top_30_words)

# Convert the list of tuples into a Pandas DataFrame for easier plotting
df_top_30 = pd.DataFrame(top_30_words, columns=['word', 'frequency'])

#bar plot
plt.figure(figsize=(10, 5))
plt.bar(df_top_30['word'], df_top_30['frequency'])

plt.title('Top 30 Most Frequent Words in a Negative Movie Review')
plt.xlabel('Word')
plt.ylabel('Frequency')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Adjust layout
plt.tight_layout()

plt.show()

print("\nThe bar plot above visualizes the frequencies of the top 30 most frequent words after preprocessing.")
```

#### Punctuation Removal Methods: Question 2 vs. Question 3(for my kind information: written version of the concept learned)

We used slightly different approaches to handle punctuation in Question 2 (comparing Inaugural Addresses with Word Clouds) and Question 3 (analyzing a Movie Review for word frequency). Clarify why.

Method used in Question 2:

In Question 2, after tokenization using word_tokenize, we had a list of tokens. The code for filtering punctuation involved a check like this (simplified):

cleaned_words = [word.lower() for word in tokens if word not in string.punctuation]

***How it works:*** This approach iterates through the list of tokens. For each word (token), it checks if the entire token is present in the string.punctuation set (which contains individual punctuation characters like '.', ',', '!', etc.).

***What it removes:*** This method will remove tokens that consist solely of punctuation marks (like a token that is just '.' or ',').

***What it doesn't remove:*** It will not remove punctuation marks that are attached to words (e.g., if "hello!" is tokenized as a single token "hello!", or if "word." is tokenized as "word."). The .lower() is then applied to these tokens, potentially leaving punctuation attached to the lowercase word (e.g., "hello!", "word.").

Method used in Question 3:

In Question 3, inside the for loop used for preprocessing, we used this line:

cleaned_token = token.translate(str.maketrans('', '', string.punctuation)).lower()

***How it works:*** This approach is more granular. It operates on a character level for each token.

str.maketrans('', '', string.punctuation): This creates a translation table that maps every punctuation character defined in string.punctuation to None, effectively marking them for deletion.

token.translate(...): This string method applies the translation table to the token string. It removes all the characters that were mapped to None (i.e., all punctuation characters).

.lower(): This converts the resulting string (after punctuation removal) to lowercase.

***What it removes:*** This method will remove punctuation characters from anywhere within or at the ends of a token. So, "hello!" becomes "hello", "word." becomes "word", and a token like "." is removed entirely.

***Why the difference in Question 3?***

The primary reason for using the translate method in Question 3 was to prepare the text more thoroughly for lemmatization. Lemmatization (reducing words to their base or dictionary form, like "running", "runs", "ran" -> "run") works best on clean words without attached punctuation. If punctuation remains, the lemmatizer might not correctly identify the base form of the word.

Using token.translate(str.maketrans('', '', string.punctuation)) ensures that punctuation characters are stripped from the words themselves (e.g., "running." becomes "running") before lemmatization is applied, leading to more **accurate** results for the frequency analysis.

In Question 2, since the main goal was a visual word cloud and not a subsequent linguistic analysis task like lemmatization, the simpler method of filtering out tokens that were only punctuation was considered sufficient.

In essence, the translate method with str.maketrans provides a more precise, character-level removal of punctuation, which was necessary for the preprocessing pipeline in Question 3 due to the inclusion of lemmatization.

### N-grams
An n-gram is a contiguous sequence of n items from a given sample of text or speech. The items can be characters, syllables, words, or even base pairs depending on the application.
In the context of text analysis, we usually talk about word n-grams.
If n = 1, it's a unigram (a single word).
If n = 2, it's a bigram (a pair of consecutive words, like "quick brown").
If n = 3, it's a trigram (a sequence of three consecutive words, like "the quick brown").
And so on for larger values of n.
N-grams are used in various NLP tasks, such as language modeling (predicting the next word in a sequence), text generation, and capturing local word dependencies. In Question 4, we'll be specifically looking at trigrams.

### Co-occurrence Matrix
A co-occurrence matrix is a matrix (a grid or table) that shows how often pairs of items appear together within a specified context.
In the context of text and word analysis, a word co-occurrence matrix shows how often two words appear together.
The "context" can be defined in different ways: it could be within the same document, within the same sentence, or within a small window of words (like an n-gram).
The matrix typically has words as both row and column headers. The value in a cell at the intersection of row 'Word A' and column 'Word B' represents the number of times Word A and Word B co-occurred within the defined context. Since co-occurrence is usually symmetrical (if A co-occurs with B, B also co-occurs with A), these matrices are often symmetrical.

### Question 4: Trigrams and Co-occurrence Matrix

This question focuses on analyzing a text from the State of the Union corpus, identifying frequent words, extracting trigrams, and building a co-occurrence matrix for the top words based on their appearance within the same trigrams.

```python
import nltk
from nltk.corpus import state_union, stopwords
from nltk.tokenize import word_tokenize
import string
from nltk import FreqDist
import pandas as pd
import numpy as np
from itertools import combinations


file_ids = state_union.fileids()
print("State of the Union Corpus File IDs (first 10):")
print(file_ids[:10])


selected_file_id = file_ids[0]
print(f"\nSelected File ID: {selected_file_id}")

raw_text = state_union.raw(selected_file_id)
print("\nFirst 500 characters of the raw text:")
print(raw_text[:500])


#Preprocess the text
text_lower = raw_text.lower()
tokens = word_tokenize(text_lower)
cleaned_tokens = [word for word in tokens if word not in string.punctuation]
stop_words = set(stopwords.words('english'))
words_cleaned = [word for word in cleaned_tokens if word not in stop_words]
print("\nFirst 20 cleaned words after removing punctuation and stopwords:")
print(words_cleaned[:20])



fdist = FreqDist(words_cleaned)
top_10_words = [word for word, freq in fdist.most_common(10)]
print("\nTop 10 Most Frequent Words:")
print(top_10_words)


# Generate trigrams from the cleaned words
trigrams = list(nltk.trigrams(words_cleaned))
print("\nFirst 10 trigrams:")
print(trigrams[:10])


co_occurrence_matrix = pd.DataFrame(0, index=top_10_words, columns=top_10_words)


for trigram in trigrams:
    words_in_top_10 = [word for word in trigram if word in top_10_words]

    # If there are at least two top 10 words in the trigram, record co-occurrences. Tip: This tri-gram is our context window of size = 3
    if len(words_in_top_10) >= 2:
        for word1, word2 in combinations(words_in_top_10, 2):
            co_occurrence_matrix.loc[word1, word2] += 1
            co_occurrence_matrix.loc[word2, word1] += 1 # For symmetry

print("\nCo-occurrence Matrix for Top 10 Words (within Trigrams):")
display(co_occurrence_matrix)
print("\nThe matrix above shows how many times each pair of the top 10 words appeared together within the same trigram in the selected text.")
```

### Question 5: Vowel Bigram Frequency in French UDHR Text

Let's analyze the frequency of consecutive vowel pairs (bigrams) within words in the French text of the Universal Declaration of Human Rights, using the correct file ID and a simplified loading style.

```python
import nltk
from nltk.corpus import udhr
from nltk import FreqDist
import string


french_file_id = 'French_Francais-Latin1'
french_text = udhr.raw(french_file_id)
print('Sample Text:',french_text[0:200])
vowels = set('aeiouyàâäèéêëîïôöùûü')  #Six written vowels + common accented vowels in French
vowel_sequences = []

#tokenization by splitting on whitespace and removing punctuation from ends
words = french_text.lower().split()
words = [word.strip(string.punctuation) for word in words if word.strip(string.punctuation)] # Remove punctuation from the beginning and end of words

for word in words:
    current_vowel_sequence = ""
    for char in word:
        if char in vowels:
            current_vowel_sequence += char
        else:
            if current_vowel_sequence:
                vowel_sequences.append(current_vowel_sequence)
            current_vowel_sequence = ""

    if current_vowel_sequence:
        vowel_sequences.append(current_vowel_sequence)

vowel_bigrams = []
for sequence in vowel_sequences:
    # Generate bigrams from each vowel sequence
    if len(sequence) > 1:
        for i in range(len(sequence) - 1):
            vowel_bigrams.append((sequence[i], sequence[i+1]))


fdist_vowel_bigrams = FreqDist(vowel_bigrams)
print("\nFrequency Distribution of Vowel Bigrams (Sorted):")
for bigram, frequency in sorted(fdist_vowel_bigrams.items(), key=lambda i: i[1], reverse=True):
    bigram_str = ''.join(bigram)
    print(f"{bigram_str}: {frequency}")



print("\nThe list above shows the frequency of consecutive vowel pairs found within words in the French UDHR text.")
```
